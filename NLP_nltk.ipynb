```python
#cleaning texts

import re
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

ps=PorterStemmer()
wordnet=WordNetLemmatizer()

sentences=nltk.sent_tokenize(corpus)

document=[]

for i in range(len(sentences)):
  review=re.sub('[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
  review=review.split()
  review=[wordnet.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review=''.join(review)
  document.append(review)


#creating a Bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(document).toarray()
```

```python
#using spacy for BOW
import re
import spacy
nlp=spacy.load('en_core_web_sm')

doc=nlp(corpus)

spacy_doc=[]
for i,sentence in enumerate(doc.sents):
  review=re.sub('[^a-zA-Z]',' ',sentence.text)
  review=review.lower()
  review=nlp(review)
  review=[token.lemma_ for token in review if not token.is_stop]
  review=" ".join(review)
  spacy_doc.append(review)


from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(spacy_doc).toarray()
```

```python
#TF-IDF
import re
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

wordnet=WordNetLemmatizer()

sentences=nltk.sent_tokenize(corpus)

document=[]

for i in range(len(sentences)):
  review=re.sub('[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
  review=review.split()
  review=[wordnet.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review=' '.join(review)
  document.append(review)


from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer()
x=cv.fit_transform(document).toarray()
```
